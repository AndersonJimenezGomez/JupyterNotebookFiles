```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = [1,2,3,4,5,6,7]
y = [50,51,42,48,49,47,45]
```

```python
plt.plot(x,y)
plt.xlabel('Day', fontsize = 20)
plt.ylabel('Temperature', fontsize = 20)
plt.title('Weather')
```

```python
plt.plot(x,y,marker='D', linestyle='',markersize=10)
plt.xlabel('Day', fontsize = 20)
plt.ylabel('Temperature', fontsize = 20)
plt.title('Weather')
```

```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.plot(days, max_t, label = 'Max')
plt.plot(days, min_t, label = 'Min',color = 'red')
plt.plot(days, avg_t, label = 'Avg')

plt.legend(loc='best', shadow=True)
plt.grid()
```

```python
days=['M','T','W','T', 'F', 'S', 'Sun']
plt.bar(days,avg_t)
```
