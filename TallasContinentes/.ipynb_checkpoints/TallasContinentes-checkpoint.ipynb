```python
#We import the libraries
import pandas as pd
import matplotlib as mtp
import numpy as np
#We import the library who's going to make the prediction
from sklearn import tree
```

```python

data = pd.read_csv('TallasContinentes.csv')
data = data.rename(columns={
    'C':'Continent',
    'Sexo': 'Gender',
    'Altura': 'Height',
    'Peso': 'Weight',
    'IMC': 'BMI'
                           })
data
```

```python
data.describe()
```

```python
clf = tree.DecisionTreeClassifier()
#continent = np.asarray(data[["Continent"]])


#x = np.array([data.Height, data.Weight, data.BMI])
gender = data.Gender

new_data = data[['Height','Weight','BMI']]
new_data = new_data.values

clf = clf.fit(new_data,gender)
```

```python
height = input("height:")
weight = input("weight:")
bmi = weight/height**2
prediction = clf.predict([[height,weight,bmi]])
print "It's very probable the person is "prediction 
```
